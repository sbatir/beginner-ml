```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 

#Load our image
image = cv2.imread('starbunnyy.jpg', cv2.IMREAD_GRAYSCALE)

#Declare our kernel as a 3x3 filter box
kernel = np.ones((3,3), np.float32)/9

#Apply Convolution 
filtered = cv2.filter2D(image, -1, kernel)

plt.imshow(filtered)
plt.title('filtered starbunny')
plt.show()
```

```python
from numpy.fft import fft2, fftshift

#Compute our 2D Fourier from the image. 
transform = fft2(image)
shifted = fftshift(transform)

#magnitude spectrum 
magnitude_spectrum = np.log(np.abs(shifted)+1)

#Display Starbunny post-FFT

plt.imshow(magnitude_spectrum)
plt.title('Magnitude Spectrum of Starbunny')
plt.show()
```

```python
import pywt

#Apply Wavelet Transform 

coeffs = pywt.dwt2(image, 'haar')
CA, (cH, cV, cD) = coeffs

plt.imshow(CA)
plt.title('Approximation')
plt.show()

plt.imshow(cH)
plt.title('Horizontal Detail')
plt.show()

plt.imshow(cV)
plt.title('Vertical Detail')
plt.show()

plt.imshow(cD)
plt.title('Diagonal Detail')
plt.show()


```
