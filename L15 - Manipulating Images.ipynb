```python
import matplotlib.pyplot as plt
import cv2

# Load an image
image = cv2.imread('starbunnyy.jpg')

# Resize by setting dimensions
resized = cv2.resize(image, (300, 200))

# Displaying original and resized images
plt.imshow(image)
plt.title('original starbunny')
plt.show()

plt.imshow(resized)
plt.title('resized starbunny')
plt.show()
```

```python
# Cropping an image [y1:y2, x1:x2]
cropped = image[50:250, 60:260]


plt.imshow(cropped)
plt.title('cropped starbunny')
plt.show()
```

```python
# Rotating an image
center = (image.shape[1] // 2, image.shape[0] // 2)
matrix = cv2.getRotationMatrix2D(center, 45, 1)  # 45-degree rotation
rotated = cv2.warpAffine(image, matrix, (image.shape[1], image.shape[0]))

plt.imshow(rotated)
plt.title('rotated starbunny')
plt.show()
```

```python
# Drawing a rectangle
cv2.rectangle(image, (50, 50), (250, 250), (0, 255, 0), 2)

# Drawing a circle
cv2.circle(image, (150, 150), 50, (0, 0, 255), -1)  # -1 fills the circle

plt.imshow(image)
plt.title('starbunny with shapes')
plt.show()
```

```python
# Adding text to an image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image, 'OpenCV', (50, 50), font, 1, (0, 255, 0), 2, cv2.LINE_AA)

# Displaying image with text

plt.imshow(image)
plt.title('starbunny with Text')
plt.show()
```
