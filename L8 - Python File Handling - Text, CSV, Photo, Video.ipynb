```python
#Reading and Writing .Txt 
# Reading from a scroll
with open('scroll.txt', 'r') as scroll:
    secrets = scroll.read()
    print(secrets)

# Writing to a scroll
with open('scroll.txt', 'w') as scroll:
    scroll.write("New secrets unveiled!")
    
print(secrets.split('\n')[:5])
```

```python
#Reading and Writing a CSV

# Reading a tome
with open('f1k_2023_l8.csv', 'r') as tome:
    reader = csv.reader(tome)
    for row in reader:
        print(row)

reader.head()

# Writing to a tome
data = [["Name", "Ability"], ["Demerzel", "Immortality"], ["Gaal", "Precognition"]]
with open('new_tome.csv', 'w') as tome:
    writer = csv.writer(tome)
    writer.writerows(data)
    
writer.head()
```

```python
#Reading and Writing a JSON 

# Reading an encyclopedia
with open('grimoire_l8.json', 'r') as grimoire:
    knowledge = json.load(grimoire)
    print(knowledge)

# Writing to an encyclopedia
knowledge = {"mermaid": "half humanoid, half fish", "centaur": "half humanoid, half horse"}
with open('new_grimoire.json', 'w') as book:
    json.dump(knowledge, book)

#Convert JSON to DF, then Preview the Head 
df = pd.DataFrame.from_dict(knowledge, orient='index')
df.head()
```
